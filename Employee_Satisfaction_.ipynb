```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("employee_survey.csv")
```

```python
stats = df.describe().T
stats['variance'] = stats['std'] ** 2
stats[['mean', 'variance']]
```

```python
correlation_matrix = df.corr()
correlation_matrix['Overall_Satisfaction'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Employee Satisfaction Correlation Heatmap")
plt.tight_layout()
plt.savefig("employee_satisfaction_heatmap.png")
plt.show()
```
